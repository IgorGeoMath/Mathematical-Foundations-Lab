```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2 # Biblioteca para visualizar conjuntos
import numpy as np

print("Laboratório de Fundamentos Matemáticos Inicializado.")
```

Na aula de Teoria dos Conjuntos, aprendemos que a Interseção ($\cap$) representa elementos comuns entre conjuntos. Em Análise de Dados, isso é fundamental para cruzar bases de dados (como o comando INNER JOIN em SQL ou ao método .merge() do Pandas.). Abaixo, aplico esse conceito para encontrar clientes recorrentes.

```python
# Conjunto A: Clientes que compraram Notebooks
# Conjunto B: Clientes que compraram Mouses
clientes_notebook = {101, 102, 103, 104, 105, 108}
clientes_mouse = {104, 105, 106, 107, 108, 109}

# 1. Interseção (A ∩ B): Quem comprou AMBOS?
# Na matemática: x ∈ A e x ∈ B
cross_sell_target = clientes_notebook.intersection(clientes_mouse)

# 2. Diferença (A - B): Quem comprou Notebook mas NÃO comprou mouse?
# Na matemática: x ∈ A e x ∉ B
oportunidade_venda = clientes_notebook.difference(clientes_mouse)

# 3. Índice de Jaccard (Métrica de Similaridade usada em Data Science)
# É a razão entre a Interseção e a União.
uniao = clientes_notebook.union(clientes_mouse)
jaccard_index = len(cross_sell_target) / len(uniao)

print(f"Clientes Alvo (Interseção): {cross_sell_target}")
print(f"Oportunidade de Venda (A - B): {oportunidade_venda}")
print(f"Similaridade dos Grupos (Jaccard): {jaccard_index:.2f}")

# Visualização Científica (Diagrama de Venn)
plt.figure(figsize=(8, 8))
venn2([clientes_notebook, clientes_mouse], set_labels=('Notebooks', 'Mouses'))
plt.title("Visualização de Conjuntos de Dados (Venn Diagram)")
plt.show()
```

O Princípio da Casa dos Pombos (Pigeonhole Principle) afirma que se $n > m$, haverá repetição. Em Cibersegurança, isso prova matematicamente a inevitabilidade das Colisões de Hash quando o espaço de chaves é menor que o de entradas.

```python
# Baseado no "Princípio da casa dos pombos"
# A aula diz: Se n > m, pelo menos uma casa terá mais de 1 pombo.
# Vamos provar isso computacionalmente com uma simulação de Hashing.

def simulacao_pombos(n_pombos, m_casas):
    casas = [0] * m_casas # Cria m casas vazias

    # Distribui os pombos aleatoriamente
    for _ in range(n_pombos):
        casa_escolhida = np.random.randint(0, m_casas)
        casas[casa_escolhida] += 1

    # Verifica colisões (Casas com > 1 pombo)
    colisoes = [i for i, x in enumerate(casas) if x > 1]
    return len(colisoes) > 0, casas

# Teste: 10 pombos em 9 casas
tem_colisao, distribuicao = simulacao_pombos(10, 9)

print(f"Simulação (10 itens em 9 slots):")
print(f"Distribuição nos slots: {distribuicao}")
print(f"Houve colisão (Princípio provado)? {'SIM' if tem_colisao else 'NÃO'}")

# Aplicação em T.i: Isso explica por que colisões de Hash são inevitáveis
# se o espaço de chaves for menor que o número de inputs.
```
